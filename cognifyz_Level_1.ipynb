```python
dataset['Restaurant Name']=dataset['Restaurant Name'].str.replace('[^a-zA-Z0-9/s]','',regex=True)
dataset.tail()
```

```python
columns_contaminated=['Restaurant Name','City','Address','Locality','Locality Verbose']
for column in columns_contaminated:
  dataset[column]=dataset[column].str.replace('[^a-zA-Z0-9/s]','',regex=True)
dataset.tail()
```

**Task: Geospatial Analysis**

```python
fig=px.scatter_mapbox(dataset,lat='Latitude',lon='Longitude',hover_name='Restaurant Name',zoom=10,)
fig.update_layout(mapbox_style="carto-positron",mapbox_zoom=10,mapbox_center_lat=dataset['Latitude'].mean(),mapbox_center_lon=dataset['Longitude'].mean(),margin=dict(l=0,r=0,t=0,b=0),)
```

```python
city_distribution=dataset['City'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=dataset,x='City',order=city_distribution.index,palette='viridis')
plt.title('Distribution of Restaurants accross Cities')
plt.xlabel('City')
plt.ylabel("Count")
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
hyderabad_restaurants=dataset[dataset['City']=='Hyderabad']
plt.figure(figsize=(8,6))
sns.countplot(data=hyderabad_restaurants,x='Locality',palette='viridis')
plt.title('Distribution in Hyderabad')
plt.xlabel('Locality')
plt.ylabel('Count')
plt.xticks(rotation=45,ha='right')
plt.show()
```
